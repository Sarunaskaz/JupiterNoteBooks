```python
# Nuskaitykite aruodas.csv failą
# Pridėkite prie DataFrame'o naują stulpelį "KainaS", kur būtų skaičiai (float'ai) iš stulpelio "€/S"
# tam panaudokite savo sukurtą funkciją, kuri tekstą verstų į float'us
# bei .apply() metodą
# Suraskite vidutinę kv. m. kainą.

aruodas = pd.read_csv('NewData/aruodas.csv')

def converter(sk):
    skaicius = float(sk[:-4])
    return skaicius

aruodas['KainaS'] = aruodas['€/S'].apply(converter)
kainu_vid = aruodas['KainaS'].mean()
print(f'Kainu Vidurkis:{kainu_vid}')


#Dar su aruodas.csv:
# Suraskite, kiek buvo pigių, kiek buvo brangių butų (mažiau už vidutinę
# kainą, daugiau nei vidutinė kaina)
#Suraskite vidutinį butų plotą.

kainos1 = []

for x in aruodas['KainaS'].values:
    if x < kainu_vid:
        kainos1.append('pigus')
    else:
        kainos1.append('brangus')

print(f'Pigus: {kainos1.count('pigus')}''\n' f'Brangus: {kainos1.count('brangus')}')

# aruodas['Plotas1'] = aruodas['Plotas'].apply(converter)
Plotu_vid = aruodas['Plotas'].mean()
print('Vidutinis plotas:', Plotu_vid)
aruodas.head()

```

```python
#AUTO.CSV failas:
#Raskite vidutinę kainą, ridą visiems automobiliams bei atskirai šiems: BMW, Audi, Toyota, Vokswagen, Ford.
#Suskirstykite automobilius pagal santykį su vidutine kaina:
# < 0.75 avg - pigūs
# 0.75 iki 1.15 avg - vidutiniai, 
# > 1.15 avg - brangūs
# Kiek kiekvienoje grupėje buvo tokių auto?
#tai vizualizuokite su bar 

auto= pd.read_csv('NewData/auto.csv', sep= "|")

# auto.head()

def converter1(sk):
    skaicius = float(sk[0:-1].replace(' ','').replace('k','').replace('Nenurodyt','0'))
    return skaicius

auto['Kainos'] = auto['price'].apply(converter1)
kainu_vid = auto['Kainos'].mean()


auto['Ridos'] = auto['rida'].apply(converter1)
ridu_vid = auto['Ridos'].mean()


print(f'Kainu vidurkis: {kainu_vid}  Ridu vidurkis: {ridu_vid}')

m_tipai = ['BMW', 'Audi', 'Toyota', 'Volkswagen', 'Ford']

for marke in m_tipai:
    skirtingu_m_tipai = auto.query('gamintojas == @marke')['Ridos'].mean()
    print(f'Marke: {marke}''\n' f'Rida: {skirtingu_m_tipai}')

print('------------')
#Suskirstykite automobilius pagal santykį su vidutine kaina:
# < 0.75 avg - pigūs
# 0.75 iki 1.15 avg - vidutiniai, 
# > 1.15 avg - brangūs
# Kiek kiekvienoje grupėje buvo tokių auto?
#tai vizualizuokite su bar 

kainos = []

for x in auto['Kainos'].values:
    if x > 0.75 * kainu_vid  and x < 1.15* kainu_vid :   
        kainos.append("vidutine")
    elif x > kainu_vid *0.75:
        
        kainos.append("pigus")
    else:
        
        kainos.append("brangus")


pigus = kainos.count('pigus')
vidutinei = kainos.count('vidutine')
brangus = kainos.count('brangus')

print("Pigus", pigus,"vidutine", vidutinei, "brangus", brangus)


# months = nypd_small['MonthNr'].value_counts().sort_index(ascending=True).index


fig, ax = plt.subplots()


ax.bar(1, pigus, color='green', label="Pigus")
ax.bar(2, vidutinei, color='red', label="Vidutinei")
ax.bar(3, brangus, color='xkcd:blue grey', label="Brangus")

ax.legend(loc='best')

plt.show()

# Kai braizau lentele x asi galiu susigalvoti pats (Priklauso kiek man reikia stulpeliu)
```
